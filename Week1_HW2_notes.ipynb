#### map()

Use map to apply the function to each element in the list
```
map(fun, iter)
```
* returns a *map object*(which is an iterator) of the results after applying the given ***function*** (equation, lambda, list, created function, etc) to each item of a given ***iterable*** (list, tuple etc.)
* The returned value from `map()` ***(map object)*** then can be passed to functions like `list()` (to create a list), `set()` (to create a set)

````python
def addition(n):
  return n+n

numbers = (1,2,3,4)
result = map(addition, numbers)
print(result)
print(list(result))
````

````python
# Double all numbers using map and lambda

numbers = (1, 2, 3, 4)
result = map(lambda x: x + x, numbers)
print(list(result))
````

````python
# Add two lists using map and lambda

numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

result = map(lambda x, y: x + y, numbers1, numbers2)
print(list(result))
````

````python
# List of strings
l = ['happy', 'cat', 'chonk']

# map() can listify the list of strings individually
test = list(map(list, l))
print(test)
````

#### apply 함수
```
apply(fun, axis)
```
함수를 column(axis=0)으로 적용할지 행(axis=1)으로 적용할지

````python
A = [5,10,15,20]
B = [100,100,100,100]
df = pd.DataFrame({'A': A, 'B':B})
df
````

````python
## sum of each rows
df2 = df.copy()
df2['sum'] = df2.apply(np.sum, axis=1)
df2
````

if로 특정 열과 행에만 함수 apply 시키기

````python
# A, B 열 데이터만 제곱값으로 변환
df3 = df2.apply(lambda x: np.square(x) if x.name in ['A','B'] else x)
df3
````

````python
# double A if it is in tens
# 특정 조건(in tens)에 맞는 행(A)에 함수(double) 적용하기
df4 = df.apply(lambda x: 2*x['A'] if (x['A']%10)==0 else x, axis=1)
df4
````
