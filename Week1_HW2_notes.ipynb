### Dataframe, list, dictionary, numpy dnarray 상호변환

```python
import numpy as np
```

```python
col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print(array1.shape)
```

```python
# 리스트를 이용해 Dataframe 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print(df_list1)
```

```python
# numpy ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print(df_array1)
```

```python
col_name2 = ['col1', 'col2', 'col3']
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print(array2.shape)
```

```python
# 2차원 리스트를 이용해 Dataframe 생성
df_list2 = pd.DataFrame(list2, columns=col_name2)
print(df_list2)
```

```python
# 2차원 numpy ndarray를 이용해 DataFrame 생성
df_array2 = pd.DataFrame(array2, columns=col_name2)
print(df_array2)
```

```python
# 딕셔너리로 만든 DataFrame
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print(df_dict)
```

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print(type(array3), array3.shape)
print(array3)
```

```python
# DataFrame -> list
list3 = df_dict.values.tolist()
print(list3)

# DataFrame -> dictionary
dict3 = df_dict.to_dict('list')
print(dict3)
```

DataFrame 및 Series에 `reset_index()`를 수행하면 새롭게 index를 연속 숫자 형으로 할당하며 원래의 index는 새로운 칼럼 'index'로 추가된다.

```python
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
```

Series에 reset_index()를 하면 index 칼럼이 추가되기 때문에 Series가 아닌 DataFrame이 반환됨

```python
print('#### A Series before reset_index ####')
a = titanic_df['Pclass'].value_counts()
print(a)
print(type(a))


print('#### A Series after reset_index ####')
b = a.reset_index(inplace=False)
print(b)
print(type(b))
```
