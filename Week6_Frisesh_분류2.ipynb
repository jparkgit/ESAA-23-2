# 3. 보팅

````python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
````

````python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
````

````python
# 데이터 확인
train.head()
````

```
pandas.get_dummies(data, prefix=None, prefix_sep='_',
                  dummy_na=False, columns=None,
                  sparse=False, drop_first=False,
                  dtype=None)
```
[pandas](https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html)

카테고리형 변수를 dummpy/indicator 변수로 변환해줌

카테고리 별 필요한 만큼 새로운 행이 생성되어 그 카테고리 유형에 따라 0,1이 입력된다

우리 데이터에서 카테고리 변수는 'type' - 유형은 red와 white이다.

`pandas.get_dummies` 적용한 뒤의 train은 type 열이 빠지고 대신 0과 1로 채워진 type_red, type_white 가 생김

````python
# 원핫 인코딩 (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
train_one.head()
````

````python
# 모델 정의 (LightGBM, XGBoost, Random Forest)
mod_lgbm = LGBMClassifier(random_state=0)
mod_xgb = XGBClassifier(random_state=0)
mod_rf = RandomForestClassifier(random_state=0)

# VotingClassifier 정의
vo_clf = VotingClassifier(estimators=[('LGBM', mod_lgbm), ('XGB', mod_xgb), ('RF', mod_rf)], voting='soft')
````

````python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
X = train_one.drop(['quality'], axis=1)
# y 는 train의 qulity 변수
y = train_one['quality']

# fit 메소드를 이용해 모델 학습
vo_clf.fit(X,y)
````

````python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
pred = vo_clf.predict(test_one)
pred
````

````python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
vo_sub['quality'] = pred
````

````python
# 정답파일 내보내기
vo_sub.to_csv('vo_pred.csv',index = False)
````
